```python
# Fill missing conversions with 0
df["total_conversion"] = df["total_conversion"].fillna(0)
df["approved_conversion"] = df["approved_conversion"].fillna(0)
```

```python
# Create Calculated Metrics
df["CTR"] = (df["clicks"] / df["impressions"]) * 100
df["CPC"] = df["spent"] / df["clicks"].replace(0, pd.NA)
df["CPA"] = df["spent"] / df["total_conversion"].replace(0, pd.NA)
```

```python
# Assume each approved conversion = $100 revenue
df["Revenue"] = df["approved_conversion"] * 100
df["ROI"] = ((df["Revenue"] - df["spent"]) / df["spent"]) * 100
```

```python
# Campaign Summary
summary = {
    "Total Spend ($)": df["spent"].sum(),
    "Total Impressions": df["impressions"].sum(),
    "Total Clicks": df["clicks"].sum(),
    "Total Conversions": df["total_conversion"].sum(),
    "Approved Conversions": df["approved_conversion"].sum(),
    "Average CTR (%)": df["CTR"].mean(),
    "Average CPC ($)": df["CPC"].mean(),
    "Average CPA ($)": df["CPA"].mean(),
    "Overall ROI (%)": df["ROI"].mean()
}

print("\n=== Campaign Summary ===")
for k, v in summary.items():
    print(f"{k}: {round(v,2)}")
```

```python
# ROI by Age
roi_age = df.groupby("age")["ROI"].mean().reset_index()
print("\n=== ROI by Age Group ===")
print(roi_age)
```

```python
# ROI by Gender
roi_gender = df.groupby("gender")["ROI"].mean().reset_index()
print("\n=== ROI by Gender ===")
print(roi_gender)
```

```python
# Top Campaigns by ROI
roi_campaign = df.groupby("fb_campaign_id")["ROI"].mean().reset_index().sort_values(by="ROI", ascending=False)
print("\n=== Top 5 Campaigns by ROI ===")
print(roi_campaign.head(5))
```
